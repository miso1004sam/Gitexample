```python
#Iterative deepening serach for solving 8-puzzle
#Han Seokhoon 3035144776

import sys

#Define puzzle state node
    
class State:
    
    def __init__(self,x=0):
        self.depth = x
        self.puzzle = []
        self.child = []
        self.parent = None
        self.pastmove = None

#function for returning zero position        
        
    def checkZero(self):
        return self.puzzle.index(0)

#function for checking the answer    
    
    def checkAnswer(self):
        if self.puzzle == [0,1,2,3,4,5,6,7,8]:
            return True
        return False

#Functions for swiping zero   
    
    def swap(self,x,y):
        temp = self.puzzle[x]
        self.puzzle[x] = self.puzzle[y]
        self.puzzle[y] = temp
    
    def zeroUp(self):
        self.swap(self.checkZero(),self.checkZero()-3)
        self.pastmove = "down"
    
    def zeroDown(self):
        self.swap(self.checkZero(),self.checkZero()+3)
        self.pastmove = "up"
  
    def zeroRight(self):
        self.swap(self.checkZero(),self.checkZero()+1)
        self.pastmove = "left"
         
    def zeroLeft(self):
        self.swap(self.checkZero(),self.checkZero()-1)
        self.pastmove = "right"
                
#Functions for adding child nodes with all possible moves 

    def addChild(self):
        global count
        if count>1000000:
            print("Maximum number of nodes reached")
            sys.exit()

        x = State(self.depth+1)
        x.puzzle = self.puzzle.copy()
        x.parent = self
        self.child.append(x)
        count += 1
        return x
    
    def addAllMoves(self):
        if self.checkZero() == 0:
            self.addChild().zeroDown()
            self.addChild().zeroRight()
        elif self.checkZero() == 1:
            self.addChild().zeroDown()
            self.addChild().zeroRight()
            self.addChild().zeroLeft()
        elif self.checkZero() == 2:
            self.addChild().zeroDown()
            self.addChild().zeroLeft()
        elif self.checkZero() == 3:
            self.addChild().zeroUp()
            self.addChild().zeroDown()
            self.addChild().zeroRight()
        elif self.checkZero() == 4:
            self.addChild().zeroUp()
            self.addChild().zeroDown()
            self.addChild().zeroRight()
            self.addChild().zeroLeft()
        elif self.checkZero() == 5:
            self.addChild().zeroUp()
            self.addChild().zeroDown()
            self.addChild().zeroLeft()
        elif self.checkZero() == 6:
            self.addChild().zeroUp()
            self.addChild().zeroRight()
        elif self.checkZero() == 7:
            self.addChild().zeroUp()
            self.addChild().zeroRight()
            self.addChild().zeroLeft()
        elif self.checkZero() == 8:
            self.addChild().zeroUp()
            self.addChild().zeroLeft()

            
#Function for printing past moves

    def printAnswer(self):
        if self.depth!=0:
            self.parent.printAnswer()
        if self.pastmove=="up":
            print("move",self.depth,": move tile",self.puzzle[self.checkZero()-3],"up")
        elif self.pastmove=="down":
            print("move",self.depth,": move tile",self.puzzle[self.checkZero()+3],"down")
        elif self.pastmove=="left":
            print("move",self.depth,": move tile",self.puzzle[self.checkZero()-1],"to the left")
        elif self.pastmove=="right":
            print("move",self.depth,": move tile",self.puzzle[self.checkZero()+1],"to the right")
            
            
#Depth Limited Search
    
    def DLS(self,depthPoint):
        if self.checkAnswer(): 
            print("Solution found!")
            print("Total",self.depth,"moves")
            self.printAnswer()
            return True
        if depthPoint>self.depth:
            self.addAllMoves()
            for i in range(len(self.child)):
                if self.child[i].DLS(depthPoint): return True
                #!!!!주의할 것! 앞에 true값을 recursion에서 계속 return 받아서 들고와야함!
        return False

#Iterative Deepening Search
    
    def IDS(self):
        for i in range(1000000):      #just a large value       
            if self.DLS(i): return True
        return False
```

```python
#Puzzle input

count = 1
initial = State()

print("Enter puzzle elements: ")

for i in range(0,9):
    x = int(input())
    initial.puzzle.append(x)

#Perform search

initial.IDS()
```

```python
# Python program to print DFS traversal from a given 
# given graph 
from collections import defaultdict 
  
# This class represents a directed graph using adjacency 
# list representation 
class Graph: 
  
    def __init__(self,vertices): 
  
        # No. of vertices 
        self.V = vertices 
  
        # default dictionary to store graph 
        self.graph = defaultdict(list) 
  
    # function to add an edge to graph 
    def addEdge(self,u,v): 
        self.graph[u].append(v) 
  
    # A function to perform a Depth-Limited search 
    # from given source 'src' 
    def DLS(self,src,target,maxDepth): 
  
        if src == target : return True
  
        # If reached the maximum depth, stop recursing. 
        if maxDepth <= 0 : return False
  
        # Recur for all the vertices adjacent to this vertex 
        for i in self.graph[src]: 
                if(self.DLS(i,target,maxDepth-1)): 
                    return True
        return False
  
    # IDDFS to search if target is reachable from v. 
    # It uses recursive DLS() 
    def IDDFS(self,src, target, maxDepth): 
  
        # Repeatedly depth-limit search till the 
        # maximum depth 
        for i in range(maxDepth): 
            if (self.DLS(src, target, i)): 
                return True
        return False
    
    # Create a graph given in the above diagram 
g = Graph (7); 
g.addEdge(0, 1) 
g.addEdge(0, 2) 
g.addEdge(1, 3) 
g.addEdge(1, 4) 
g.addEdge(2, 5) 
g.addEdge(2, 6) 
  
target = 6; maxDepth = 3; src = 0
  
if g.IDDFS(src, target, maxDepth) == True: 
    print ("Target is reachable from source " +
        "within max depth") 
else : 
    print ("Target is NOT reachable from source " +
        "within max depth") 
  
# This code is contributed by Neelam Pandey 
```
